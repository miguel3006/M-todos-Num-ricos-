```python
import math
import scipy as sp
import matplotlib.pyplot as plt
from numpy import *

v=40
g=9.8
m=68.1
t=10
fc=0
x1=12
x2=16
xr=0
i=0
fx=0
ft=0
x=[]
y=[]
l=[1,0,-1]
xl=[xr,xr,xr]
fh=(g*m/x1)*(1-pow(math.exp(1),-(x1/m)*t))-v
fk=(g*m/x2)*(1-pow(math.exp(1),-(x2/m)*t))-v
ft=fh*fk
fu=0
if ft<0:  
    print("              iteracion"+"                         x1"+"                          x2"+"                           xr");
    while True:
        xr=(x1+x2)/2
        fc=(g*m/x1)*(1-pow(math.exp(1),-(x1/m)*t))-v
        fx=(g*m/xr)*(1-pow(math.exp(1),-(xr/m)*t))-v
        fu=(g*m/(i+1))*(1-pow(math.exp(1),-((i+1)/m)*t))-v
        ft=fc*fx
        if ft>0:
            print ("{0:25} | {1:25} | {2:25} | {3:25} ".format((i),x1,x2,xr));
            x1=xr
        elif ft<0:
            print ("{0:25} | {1:25} | {2:25} | {3:25} ".format((i),x1,x2,xr));
            x2=xr 
        elif ft==0:
            print ("{0:25} | {1:25} | {2:25} | {3:25} ".format((i),x1,x2,xr)); 
            break
        #print(xr);
        if i>100:
            break
        i=i+1 
        x.append(i)
        y.append(fu)
xl=[xr,xr,xr]
plt.plot(xl,l,'k',linewidth=2,label='xr raiz')
plt.hold(True)
plt.plot(x,y,'r',linewidth=2,label='analitico')

plt.legend(loc = 2)
plt.show()


    
    
```
