```python
from numpy import *
import matplotlib.pyplot as plt
g= 9.8
m= 68.1
c= 12.5
t=2
b=0
b1=0
Ev=0
Er=0
Ea=0
v3=0
e=[]
d=[]
x=[]
z=[]
y=[]
p=[]
print("         tiempo  "+" forma analitica  "+"  metodos numericos  "+"  Error verdadero"+"  Error relativo"+"  Error aproximado");
for i in range(50):
    v=(g*m/c)*(1-pow(math.exp(1),-(c/m)*i*2))
    x.append(i)
    y.append(v)
#print("metodos numericos");
    if i==0:
       v1=0
    else:
        b1=b
        v1=b+(g-((c*b)/m))*(t)
        b=v1
    p.append(v1)
#print("Error verdadero");
    Ev=(v-v1)
#print("Error relativo");
    if i==0:
        Er=0
    else:
        Er=(Ev/v1)*100
#print("Error aproximado");
    if b==0:
        Ea=0
    else:
        Ea=((v1-b1)/v1)*100
    print ("{0:15} | {1:15} | {2:15} | {3:15}   | {4:15} | {5:15}".format((i*2),v,v1,Ev,Er,Ea));
    z.append(Er) 
    d.append(Ev*(-1))
    
plt.plot(x,y,'r',linewidth=2,label='analitico')
plt.hold(True)
plt.grid(True)
plt.plot(x,p,'k--',linewidth=2,label='metodos')
plt.plot(x,z,'r--',linewidth=2,label='Er')
plt.plot(x,d,'k',linewidth=2,label='Ev')
plt.legend(loc = 'lower right')
plt.title("Caida Libre")
plt.xlabel("Tiempo") 
plt.ylabel("Velocidad") 
plt.show()
```
